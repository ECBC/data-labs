#### Define data.

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)
plt.subplot(1,2,1)
plt.plot(x,y,'tab:orange')
# plt.figure(figsize=(3,8))
plt.title('Subplot X and Y')
plt.subplot(1,2,2)
plt.plot(x,z,'tab:green')
plt.title('Subplot X and Z')
```

```python
# your code here-2st way (call `subplots` only once not using the `index` parameter)
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x, y, 'tab:orange')
ax1.set_title('Subplot X and Y')
ax2.plot(x, z, 'tab:green')
ax2.set_title('Subplot X and Z')
```

#### Augmenting your previous code, resize your previous plot.

*Hint: Add the `figsize` argument in `plt.subplots()`*

```python
# your code here
fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(15,4))
ax1.plot(x, y, 'tab:orange')
ax1.set_title('Subplot X and Y')
ax2.plot(x, z, 'tab:green')
ax2.set_title('Subplot X and Z')
```

#### Augmenting your previous code, label your axes.

*Hint: call `set_xlabel` and `set_ylabel`*

```python
# your code here
fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(15,4))
ax1.plot(x, y, 'tab:orange')
ax1.set_title('Subplot X and Y')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.plot(x, z, 'tab:green')
ax2.set_xlabel('x')
ax2.set_ylabel('z')
ax2.set_title('Subplot X and Z')
```
